```python
# Importing Libraries
import pandas as pd
import numpy as np
import time
import selenium
import requests
import re
from bs4 import BeautifulSoup
from selenium import webdriver 
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.chrome.service import Service
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
import warnings
warnings.filterwarnings("ignore")
```

```python
#connect to the webdriver
driver = webdriver.Chrome("chromedriver.exe")
```

```python
url = "https://www.cardekho.com/used-cars+in+bangalore"
driver.get(url)
time.sleep(6)
```

```python
brand=[]
model=[]
variant=[]
Driven_KM=[]
fuel=[]
price=[]
```

```python
time.sleep(3)

start=0
end=4
for page in range(start,end):
    
    brands=driver.find_elements(By.XPATH,"//div[@class='gsc_col-xs-7  carsName']/a")
    for i in brands:
        brand.append(i.text)
        
    prices=driver.find_elements(By.XPATH,"//span[@class='amnt ']")
    for i in prices:
        price.append(i.text)
        
    models=driver.find_elements(By.XPATH,"//div[@class='car_variant']")
    for i in models:
        model.append(i.text)  
        
    kms=driver.find_elements(By.XPATH,("//div[@class='truncate dotlist-2']/div[1]"))
    for i in kms:
        Driven_KM.append(i.text)
        
    fuels=driver.find_elements(By.XPATH,("//div[@class='truncate dotlist-2']/div[2]"))
    for i in fuels:
        fuel.append(i.text)
        
    variants=driver.find_elements(By.XPATH,("//div[@class='truncate dotlist-2']/div[3]"))
    for i in variants:
        variant.append(i.text)
```

```python
df1=pd.DataFrame({
'Brand':brand[:],                 
'Price':price[:],
'Model':model[:],
'KMS_driven':Driven_KM[:],
'Fuel':fuel[:],
'Variant':variant[:]})
df1
```

```python
driver.close()
```
